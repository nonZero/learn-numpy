```python
n = 5
for i in range(n):
    plt.plot(x, y*i)
```

```python
n = 5
for i in range(n):
    plt.plot(x, y*i, label=f"y * {i}")
plt.legend();
```

```python
for i in range(n):
    plt.subplot(n, 1, i+1)
    plt.plot(x,y*i)
    plt.title(f"y * {i}")
```

```python
n = 6
for i in range(n):
    plt.subplot(2, 3, i+1)
    plt.plot(x,y*i)
    plt.title(f"y * {i}")
```

```python
n = 6
for i in range(n):
    plt.subplot(2, 3, i+1)
    plt.plot(x,y*i)
    plt.title(f"y * {i}")
plt.tight_layout()    
```

```python
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(12,5), sharex='all', sharey='all', facecolor="0.7")
for i, ax in enumerate(axes.flat):
    ax.plot(x,np.sin(x**i))
    ax.set_title(f"$sin(y^{i})$")
    ax.set_facecolor("xkcd:ivory")
fig.suptitle("THE FIGURE");    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
