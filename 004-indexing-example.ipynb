## Disaplying a 2d array as an image

```python
a = np.arange(100).reshape(10, 10)
```

```python
imshow(a)
```

```python
imshow(a, cmap=cm.gray)
colorbar()
```

## Indexing Example: RGB data

```python
m = np.random.random((1024, 768, 3))
```

```python
m
```

```python
imshow(m)
```

```python
m[212, 600] # one pixel
```

```python
m[100:105, 200:208] # a 5x8x3 view
```

```python
m[:,:,0] # The R channel, 1024x768 view
```

```python
imshow(m[:,:,0])
colorbar()
```

```python
imshow(m[:,:,0], cmap=cm.gray)
colorbar()
```
