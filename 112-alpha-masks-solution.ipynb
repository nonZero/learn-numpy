# Adding an Alpha (Transparency) Channel

```python
a = plt.imread("pencils.jpeg") / 256
plt.imshow(a);
```

```python
def rgb2gray(rgb):
    return rgb[...,:3] @ [0.299, 0.587, 0.114]
```

```python
b = rgb2gray(a)
```

```python
c = b[:,:,np.newaxis] * [1,1,1]
m = (a[:,:,0] > 0.1) & (a[:,:,1] < 0.1)
c[m] = a[m]
plt.imshow(c);
```

```python
c = b[:,:,np.newaxis] * [1,1,1]
m = (a[:,:,2] > 0.1) & (a[:,:,0] < 0.1)
c[m] = a[m]
plt.imshow(c);
```
