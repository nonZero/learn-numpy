# Broadcasting Into Grids

```python
a = np.array([100, 200 ,300])
b = np.array([1, 2, 3, 4, 5])
```

```python
a.shape
```

```python
b.shape
```

```python
a + b
```

```python
a[:,np.newaxis]
```

```python
a[:,np.newaxis].shape
```

```python
a[:,np.newaxis] + b
```

```python
b[:,np.newaxis]
```

```python
b[:,np.newaxis].shape
```

```python
a + b[:,np.newaxis]
```
