```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = plt.imread("pencils.jpeg") / 256
plt.imshow(a);
```

```python
plt.figure(figsize=(16,4))
for i, c in enumerate("rgb"):
    plt.subplot(1, 3, i+1)
    plt.hist(a[:,:,i].ravel(), bins=64, color=c)
    plt.title(c)
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("red")
plt.ylabel("green")
plt.hist2d(a[:,:,0].ravel(), a[:,:,1].ravel(), bins=64, cmap='hot');
```

# Adding an Alpha (Transparency) Channel

```python
y = np.ones(a.shape[0])
x = np.linspace(0, 1, a.shape[1])
alpha = np.multiply.outer(y, x)
alpha.shape
```

```python
rgba = np.dstack((a, alpha))
rgba.shape
```

```python
plt.imshow(rgba);
```

# Converting to Grayscale
* More Info: https://en.wikipedia.org/wiki/Grayscale#Converting_color_to_grayscale

```python
def rgb2gray(rgb):
    return rgb[...,:3] @ [0.299, 0.587, 0.114]
```

```python
b = rgb2gray(a)
b.shape
```

```python
plt.imshow(b, 'gray', vmin=0, vmax=1)
plt.colorbar(fraction=0.03);
```
