```python
b = np.random.random((100,150))
```

```python
plt.imshow(b, 'gray')
plt.colorbar();
```

```python
plt.imshow(np.sort(b), 'gray')
#plt.colorbar();
plt.colorbar(orientation='horizontal',fraction=0.06);
```

```python
plt.imshow(np.sort(b, axis=0), 'gray')
plt.colorbar(fraction=0.03);
```

```python
plt.imshow(np.sort(b.ravel()).reshape(b.shape), 'gray')
plt.colorbar();
```
