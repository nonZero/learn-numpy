# Jupyter Lab/Notebook Keyboard Shortcuts

* `Enter`/`Esc`: Enter/Exit edit cell mode
* `Ctrl`/`Alt`/`Shift`+`Enter`: Run cell
* `a`/`b`: Add cell above/below
* `dd`: Delete cell
* `c`/`x`/`v`: Copy/Cut/Paste cell
* `y`/`m`: Change cell to code/markdown
