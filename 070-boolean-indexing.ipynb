## Boolean Indexing

````python
a = np.array([15, 6, 22, 28, 9, 12])
a
````

To create a new array with with some values from `a` a boolean mask in the same shape of a can be used:

````python
m = np.array([False, True, True, False, False, True])
````

````python
print(a.shape, m.shape)
````

```
a = |   15   |   6   |  22   |  28   |   9   |  12   
m = |  False | True  | True  | False | False | True  
```

````python
a[m]
````

Boolean arrays (masks) can be created with comparison and logical operators:

````python
m2 = a > 10
m2
````

````python
a[m2]
````

````python
a[a > 10]
````

````python
a
````

````python
a % 2 == 0
````

````python
a[a % 2 == 0]
````

Another example of a 2d mask:

````python
b = np.arange(24).reshape(4,6)
b
````

````python
b % 5
````

````python
m3 = b % 5 < 2
m3
````

````python
c = np.random.randint(1, 100, m3.shape)
c
````

````python
c[m3]  # does not keep the shape!  returns a 1d array.
````

````python
c[c > 90] # does not keep the shape! returns a 1d array.
````

````python
c[[True, False, True, True]] # Still 2d, but the length of the first dimension is 3 - the number of Trues!  
````

````python
c[:,[False, True, True, True, False, True]] # Still 2d, but the length of the second dimension is 4 - the number of Trues!
````

Let's try making a array inluding all rows in c that the sum of the numbers in the row is bigger than 200:

````python
np.sum(c, axis=1)
````

````python
np.sum(c, axis=1) > 200
````

````python
c[np.sum(c, axis=1) > 200]
````
